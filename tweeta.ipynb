```python
import csv

keys = ["i am depressed", "i want to die", "commit suicide", "my death", "kill myself", "hang myself", "tired of life",
        "feel sorry","heart broken", "feel rejected", "feel frustrated", "feel sad", "feel lonely", "i will be gone",
        "end my life"]

unreal = ["just kidding", "playing", "not serious", "haha", "hahaha", "messing with you", "jokes", "lol", "lmao", "dwl",
          "lmfao", "guy", "guys", "someone", "he", "she", "they"]

alert = 0

with open("tweets/tweets.csv", "r", errors='ignore') as suicide:
    reader = csv.reader(suicide, delimiter=',')
    for lines in reader:
        line = lines[10].lower()
        for i in keys:
            if i in line:
                print(line)
                print(c.Username+" is being followed up...")
                alert += 1
                for k in unreal:
                    if k in line:
                        print(line)
                        print("Oops! "+c.Username+" might just be kidding...")
                        alert -= 1
                        break
        if alert > 3:
            print(c.Username+" is likely to commit suicide!")
            break
```
